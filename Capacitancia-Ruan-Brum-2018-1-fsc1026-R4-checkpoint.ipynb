```python
#HANSO
# -*- coding: utf-8 -*-
"@autur: Helena Mota"

import numpy as np
import math as mt
import matplotlib.pyplot as plt
#Dados do problema
#- Os discos do capacitor são de alumínio e possuem um diâmentro de 80 mm;
#- Simule a capacitância para distâncias, de afastamento das placas, entre 1mm até 50mm.
#- Assuma que o meio dielétrico entre as placas seja Teflon, que possui constante dielétrica com valor 2,1.

k = 2.1                      # constante dielétrica
E = 8.85 * (10**(-12))       # contante de permissividade
A = (mt.pi) * ((.04)**2)     # area

lista = []
for valor in np.arange(.001, .05, .001):        # arange: começa em .001, vai até .05, incrementa de .001 em .001
    lista.append(valor)
#    print (lista)
#   print (valor)
d = np.array(lista)

C = (k * E * A) / d          # C = capacitancia
##print (C)                   

plt.title(u"Grafico em Funcao da capacitancia e a distancia")
plt.xlabel("Distancia")              
plt.ylabel("Capacitancia")              

plt.plot(d, C)
plt.grid()
#plt.savefig("HelenaMota.png")      
plt.show()                        
```
